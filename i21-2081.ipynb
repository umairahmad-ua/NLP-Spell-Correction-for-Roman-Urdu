#  Spell Correction for Roman Urdu


```python
import pandas as pd
import string
from collections import defaultdict  
```

#  Declaration And Intialization


```python
Vocab = defaultdict(int)
Unigrams=defaultdict(float)
dict_misspell=defaultdict(list)
Insert_Matrix = defaultdict(lambda: defaultdict(int))
Delete_Matrix = defaultdict(lambda: defaultdict(int))
Sub_Matrix = defaultdict(lambda: defaultdict(int))
Trans_Matrix = defaultdict(lambda: defaultdict(int))
Vocab_lst=list()
word_len=0
Y=list(string.ascii_lowercase)
Y.append('@')
X=list(string.ascii_lowercase)
for i in X:
    for j in Y:
        Insert_Matrix[i][j]=0
        Delete_Matrix[i][j]=0
        Sub_Matrix[i][j]=0
        Trans_Matrix[i][j]=0
```

#  File Reading and Dictionaries Intializations for Model Traning


```python
#Misspelling File Reading 
df=pd.read_csv (r'misspellings.csv')
for index, row in df.iterrows():
    dict_misspell[row['Correct']]=row[' Wrong'].split("\t")
```

```python
#Data file Reading 
file_name = "C:/Users/Umair/Desktop/NLP Assignment 4/data.txt"
txt_file = open(file_name)
for line in txt_file:
    words=line.split(" ")
    for word in words:
        if "\n" in word:
            word=word.replace('\n','')
        Vocab[word]+=1
        word_len=word_len+1
txt_file.close()
Vocab_lst=[k  for  k in  Vocab]
```

#  Unigram Model Traning and frequency calculation
- Function "cal_word_probabilty" returns the probabilty of given word p(w) in the data 
- It also save a .csv file name "unigram_model" that contains the probibilties P(w) of all the words available in the dataset


```python
#Unigram Model Traning and frequency calculation
def cal_word_probabilty(word):
    for key in Vocab:
        Unigrams[key]=Vocab[key]/word_len
    uni_gram_freq=pd.DataFrame(Unigrams.items(), columns=['Uni_Grams', 'Probabilities'])
    uni_gram_freq.to_csv(r'unigram_model.csv')
    return Unigrams[word]
uni_gram_freq
```

#  Confusion Matrix for Insert, Delete, Substitute and Transpose
- Each cell represent the each error model creations 


```python
#Insert Error Confusion Matrix
for key,words in dict_misspell.items():
    for word in words:
        for i in range(len(word)+1):
            for j in X:
                temp=word[:i] + j + word[i:]
                if(temp==key):
                    if i==0:
                        temp=temp+'@'
                    Insert_Matrix[j][temp[i-1]]+=1 #j==w, temp[i]==x
pd.DataFrame(Insert_Matrix)
```

```python
#Delete Error Confusion Matrix
for key,words in dict_misspell.items():
    for word in words:
        for i in range(len(word)):
            temp=word[:i] + word[i+1:]
            if(temp==key):
                if i>=len(temp)-1:
                    temp=temp+'@'
                    word=word+'@'
                Delete_Matrix[temp[i]][word[i]]+=1 #w=temp[i]  , x=word[i]
pd.DataFrame(Delete_Matrix)
```

```python
#Substitution Error Confusion Matrix
for key,words in dict_misspell.items():
    for word in words:
        for i in range(len(word)):
            for j in X:
                temp=word[:i] + j + word[i+1:]
                if(temp==key):
                    Sub_Matrix[j][word[i]]+=1    #w=j  ,  x=word[i]
pd.DataFrame(Sub_Matrix)
```

```python
#Transpose Error Confusion Matrix
for key,words in dict_misspell.items():
    for word in words:
        for i in range(len(word)-1):
            char1=word[i]
            char2=word[i+1]
            temp=word[:i] + word[i+2:]
            temp=temp[:i] + char2+char1 + temp[i:]
            if(temp==key):
                Trans_Matrix[word[i+1]][word[i]]+=1    #w=word[i+1]  ,  x=word[i]
pd.DataFrame(Trans_Matrix)
```

#  Calculate the Probability of P(x/w)
- Function "cal_error_model_probability" receive three parameters mistypes word x,intended word w and operation i.e. insert,delete,Substitute,Transpose 
- Return the probabilty of P(x/w)


```python
#Calculate the Probability of P(x|w)

def cal_error_model_probability(x,w,op):
    prob=0.0
    if(op=="insert"):
        total_occurance=sum(d[w] for d in Insert_Matrix.values() if d)
        if total_occurance!=0:
            prob=(Insert_Matrix[w][x])/total_occurance
        return prob
    elif (op=="delete"):
        total_occurance=sum(d[w] for d in Delete_Matrix.values() if d)
        if total_occurance!=0:
            prob=(Delete_Matrix[w][x])/total_occurance
        return prob
    elif (op=="sub"):
        total_occurance=sum(d[w] for d in Sub_Matrix.values() if d)
        if total_occurance!=0:
            prob=(Sub_Matrix[w][x])/total_occurance
        return prob
    elif (op=="trans"):
        total_occurance=sum(d[w] for d in Trans_Matrix.values() if d)
        if total_occurance!=0:
            prob=(Trans_Matrix[w][x])/total_occurance
        return prob
```

#  Candidate Words Generation
- Function "cal_error_model_probability" receive one parameters string word 
- Return the all possible candidates list from all four operations(insert, delete, substitute,transpose)
- Each row of list contain the four parameters (candidate word,w,x,operation)
- Sample output of "kyaa" is displayed


```python
# Compute the candidate words

def candidate_words(word):
    candidate_words_lst=[]
    parameters=[]
    
    #Insert
    for i in range(len(word)+1):
        for j in X:
            temp=word[:i] + j + word[i:]
            if temp in Vocab_lst:
                parameters.append(temp)
                if i==0:
                    temp=temp+'@'
                parameters.append(j)#w
                parameters.append(temp[i-1])#x
                parameters.append("insert")
                candidate_words_lst.append(parameters)
                parameters=[]    
                
    #Delete
    parameters=[]
    for i in range(len(word)+1):
            cand_word=word
            temp=word[:i] + word[i+1:]
            if temp in Vocab_lst:
                parameters.append(temp)
                if i>=len(temp)-1:
                    temp=temp+'@'
                    cand_word=cand_word+'@'
                parameters.append(temp[i])#w
                parameters.append(cand_word[i])#x
                parameters.append("delete")
                candidate_words_lst.append(parameters)
                parameters=[]
                          
    #Substitution
    parameters=[]
    for i in range(len(word)):
        for j in X:
            temp=word[:i] + j + word[i+1:]
            if temp in Vocab_lst:
                parameters.append(temp)
                parameters.append(j)#w
                parameters.append(word[i])#x
                parameters.append("sub")
                candidate_words_lst.append(parameters)
                parameters=[]

    #Transpose 
    parameters=[]
    for i in range(len(word)-1):
        char1=word[i]
        char2=word[i+1]
        temp=word[:i] + word[i+2:]
        temp=temp[:i] + char2+char1 + temp[i:]
        if temp in Vocab_lst:
            parameters.append(temp)
            parameters.append(word[i+1])#w
            parameters.append(word[i])#x
            parameters.append("trans")
            candidate_words_lst.append(parameters)
            parameters=[]
                   
    return candidate_words_lst  
candidate_words("kyaa")
```

#   Selection Model using argmax:
- Function "corrected_word" receive one list consisting of four parameters candidate words, w,x,operation 
- Return the most probable corrected word w for the misspelled word x.
- incase of No candidate word it will display "No Match Found"
- Sample output of "kyaa" is displayed


```python
#Most probable corrected word
def corrected_word(candidate_words):
    if candidate_words==[]:
        final_word="No Match Found"
        return final_word
    final_word=""
    maximum=0
    for word,w,x,operation in candidate_words:
        error_prob=cal_error_model_probability(x,w,operation)
        word_prob=cal_word_probabilty(word)
        prob=error_prob*word_prob
        print(str(prob)+"     "+word)
        if(prob>=maximum):
            maximum=prob
            final_word=word
    return final_word
    
print("\n\n\nFinal Correct Word= "+corrected_word(candidate_words("kyaa")))
```
